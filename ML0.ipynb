```python
A = np.arange(1, 26).reshape([5, 5])
res = np.array([ ], int)
for i in range(len(A)):
    a = A[i][1::2] if i%2 else A[i][::2]
    res = np.concatenate((res,a))
res
```

Выделите в отдельную таблицу последние десять строк. Уберите в ней столбец с ценой. Склейте ее с первоначальной таблицей при помощи append. Заполните отсутствующие значения цены средним по таблице.

```python
df = data1.iloc[-10:].drop("Цена", axis=1)
df = data1.append(df)
df.iloc[np.where(pd.isnull(df))] = df.loc[:,"Цена"].mean()
df
```

Выделите пять последних колонок в отдельную таблицу. Удалите в ней строки, в которых цена ниже среднего. Присоедините эту таблицу к изначальной (выберите самый подходящий тип соединения).

```python
df = data1.iloc[:, -5:]
price = data1.loc[:, "Цена"].mean()
for index, row in df.iterrows():
    if row["Цена"] < price:
        df.drop(index, inplace=True)
print(price)
df = df.merge(data1, how="inner")   #####посмотреть потом
df
```

Выведите таблицу, содержащую среднюю цену и количество квартир на каждом этаже из первоначального набора данных.


```python
price = data1.loc[:, ["Этаж", "Цена"]].groupby("Этаж").mean()
rooms = data1.loc[:, ["Этаж", "Кол-во комнат"]].groupby("Этаж").mean()
df = price.merge(rooms, how="inner", on="Этаж")
display(df)
```

Постройте диаграммы рассеяния для признаков Rooms, Square, HouseFloor, HouseYear в зависимости от целевой переменной Price в одной области figure. Оцените визуально, есть ли среди них такие, на которых разброс точек близок к линейной функции.

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
rooms, square, housefloor, houseyear = axs[0, 0], axs[0, 1], axs[1, 0], axs[1, 1]
rooms.scatter(x=data.Price, y=data.Rooms, s=3)
rooms.set_xlabel("Rooms")
rooms.set_ylabel("Price")

square.scatter(x=data.Price, y=data.Square, s=3)
square.set_xlabel("Square")
square.set_ylabel("Price")

housefloor.scatter(x=data.Price, y=data.HouseFloor, s=3)
housefloor.set_xlabel("HouseFloor")
housefloor.set_ylabel("Price")

houseyear.scatter(x=data.Price, y=data.HouseYear, s=3)
houseyear.set_xlabel("HouseYear")
houseyear.set_ylabel("Price")
```
